## Base Data 전처리

```python
import numpy as np
import pandas as pd

base_data = pd.read_excel('seoul_apart.xlsx').iloc[:,:9]
base_data
```

```python
base_data['전용면적'] = base_data.전용면적.apply(lambda x: np.nan if x == ' ' else x)
base_data.ffill(axis = 0, inplace = True)
base_data = base_data.drop_duplicates()
base_data.to_csv('base_data.csv',index = False)
base_data
```
