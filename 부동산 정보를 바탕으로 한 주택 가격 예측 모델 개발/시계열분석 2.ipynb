```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font', family = 'Malgun Gothic')

apart_base_data = pd.read_csv('apart_base_data.csv', encoding = 'cp949')
apart_base_data.head()
```

```python
base = apart_base_data[['단지','전용면적','거래금액','계약년도','계약일']]
base
```

```python
base['계약월'] = base['계약일'].apply(lambda x : '{0:02d}'.format(int(str(x).split('.')[0])))
base['계약일'] = base['계약일'].apply(lambda x : '{0:02d}'.format(int(str(x).split('.')[1])))
```

```python
base
```

```python
base['계약년도'] = base.계약년도.apply(lambda x : str(int(x)))
base['계약시기'] = base['계약년도'] + '-' + base['계약월'] + '-' + base['계약일']
base.head()
```

```python
base['평당가격'] = round(base.거래금액/base.전용면적,2)
base
```

```python
base['계약년도'] = base.계약년도.apply(lambda x : str(int(x)))
base['계약일'] = base['계약일'].apply(lambda x : '{0:02d}'.format(int(str(x).split('.')[0])))
base['계약시기'] =  base['계약년도'] + '-' + base['계약일']
base
```

```python
dusan = base.query('단지 == "월곡두산위브"')[['계약시기','평당가격']]
dusan
```

```python
dusan = dusan.groupby('계약시기')['평당가격'].agg(lambda x : round(x.mean(),0)).reset_index().sort_values(by = '계약시기')
dusan
```

```python
from pmdarima.arima import auto_arima
```

```python
dusan.set_index('계약시기',inplace = True)
```

```python
dusan.plot()
```

```python
from pmdarima.arima import ADFTest
adf_test = ADFTest(alpha = 0.05)
adf_test.should_diff(dusan)
```

```python
dusan[-143:]
```

```python
train = dusan[:-143]
test = dusan[-143:]
plt.plot(train)
plt.plot(test)
plt.xticks(rotation = 90)
plt.show()
```

```python
arima_model1 = auto_arima(train, start_p = 0, d = 1, start_q = 0,
                         max_p = 5, max_d = 5, max_q = 5, start_P = 0,
                         D = 1, start_Q = 0, max_P = 5, max_D = 5,
                         max_Q = 5, m = 12, seasonal = True,
                         error_action = 'warn', trace = True,
                         supress_warnings = True, stepwise = True,
                         random_state = 20, n_fits = 50)
```

```python
arima_model1.summary()
```

```python
prediction = pd.DataFrame(arima_model1.predict(n_periods = 143),index = test.index)
prediction.columns = ['predicted_price']
prediction
```

```python
plt.figure(figsize = (8,5))
plt.plot(train,label = 'Training')
plt.plot(test,label = 'Test')
plt.plot(prediction,label = 'Predicted')
plt.legend(loc = 'Left corner')
plt.show()
```

```python
from sklearn.metrics import r2_score
test['predicted_price'] = prediction
r2_score(test['평당가격'],test['predicted_price'])
```
