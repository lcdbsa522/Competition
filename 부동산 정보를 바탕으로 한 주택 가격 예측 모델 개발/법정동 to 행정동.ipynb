```python
# 사용하는 모듈 불러오기
from selenium import webdriver
import time
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.alert import Alert

# # csv 파일 만들기
# f = open('seoul_apart.csv','w')
# f.write("단지,준공년도,전용면적,계약년도,계약일,거래금액,층,자치구,법정동\n")

# 드라이버 실행하기
driver = webdriver.Chrome('./chromedriver')
driver.get('http://land.seoul.go.kr/land/klis/findHjdongAddress.do')

# 구 리스트 만들기
gu_xpath = driver.find_element_by_xpath('/html/body/div/div/div/div[2]/div[1]/form/div/div[1]/select[1]')
gu_list = gu_xpath.text.split('\n')[1:]
time.sleep(0.3)

# 구 선택하기
for gu in gu_list:
    Select(gu_xpath).select_by_visible_text(gu)
    time.sleep(0.1)
    
    # 동 리스트 만들기
    dong_xpath = driver.find_element_by_xpath('/html/body/div/div/div/div[2]/div[1]/form/div/div[1]/select[2]')
    dong_list = dong_xpath.text.split('\n')[1:]
    
    # 동 선택하기
    for dong in dong_list:
        Select(dong_xpath).select_by_visible_text(dong)
#         time.sleep(0.1)
    # 지번 넣어주기
```
