```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
%matplotlib inline

# 스케일링
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

# Data Loading

```python
# 데이터 불러오기
target = pd.read_csv('cust_train.csv', encoding = 'UTF-8')
numeric_f = pd.read_csv('numeric_features.csv', encoding = 'UTF-8')
sparse_f = pd.read_csv('sparse_pca.csv', encoding = 'UTF-8')
```

```python
# numeric과 sparse feautre merge 후 CLNT_ID 제거
features = pd.merge(numeric_f, sparse_f, on = 'CLNT_ID', how = 'left')
del features['CLNT_ID']
col_name = list(features.columns)
features
```

# Feature Scaling

```python
# 이상치 제거
features = features.apply(lambda x: x.clip(x.quantile(.05), x.quantile(.95)), axis=0)
features
```

```python
# Scaler 선언
mms = MinMaxScaler()
ss = StandardScaler()

# Scaling
features = mms.fit_transform(features)
features = ss.fit_transform(features)

# 다시 dataframe 형태로 바꾸기
features = pd.DataFrame(features, columns = [str(i) for i in col_name])
display(features)

# train과 test feature 분리하기
train_features = features.iloc[:263104,:]
test_features = features.iloc[263104:,:]

display(train_features, test_features)
```

# Feature Save

```python
train_features.to_csv('train_features.csv', encoding = 'UTF-8', index = False)
test_features.to_csv('test_features.csv', encoding = 'UTF-8', index = False)
```
